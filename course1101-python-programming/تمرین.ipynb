```python
def factorial(n):
    fact = 1
    for num in range(2, n + 1):
        fact *= num
    return fact
x=100
n=10
S=x #sum
for i in range(1,n):
  S+= (-1)**i*(x**(2*i+1))/factorial(2*i+1)
print(S)
```

```python
N=int(input('number of persons: '))
tex=[0]*N
for i in range(N):
  c=int(input("Enter the amount of income of person %d (in million tomans): " % (i) ))
  if c <= 20:
    tex[i]=0
  elif c >20 & c <= 50:
    tex[i]=0.1*(c-20)
  else:
    tex[i]=0.15*(c-50)

print(tex)
```
